```python
import cv2
import matplotlib.pyplot as plt
# path
  
# Using cv2.imread() method
img = cv2.imread("frutilla.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
# Displaying the image
cv2.imshow('origina', img)
cv2.imshow('gris', gray)

cv2.waitKey(0)
cv2.destroyAllWindows()

hist = cv2.calcHist([gray],[0],None,[256],[0,256])
  
# plot the above computed histogram
plt.plot(hist, color='b')
plt.show()
```

```python
plt.hist(gray.ravel(),256,[0,256])
plt.title('Histogram for gray scale image')
plt.show()
```

```python
ret,thresh100 = cv2.threshold(gray,100,255,cv2.THRESH_BINARY)
cv2.imshow('umbral 100', thresh100)

ret,thresh150 = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
cv2.imshow('umbral 150', thresh150)

ret,thresh50 = cv2.threshold(gray,50,255,cv2.THRESH_BINARY)
cv2.imshow('umbral 50', thresh50)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
def red(imagen,thresh1,thresh2):
    largo,ancho,canales=np.shape(imagen)

    for i in range(largo):
        for j in range(ancho):
            pixel=imagen[i,j]
            if pixel[2]<thresh1 or pixel[1]>thresh2:
                imagen[i,j]=np.array([0,0,0])
    return imagen


            
```

```python
cv2.imshow('actual', img)
import copy
copiaimg=copy.copy(img)
redimg=red(copiaimg,120,100)
cv2.imshow('nuevo rojo', redimg)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
